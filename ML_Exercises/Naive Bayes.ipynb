```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine 
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
wine.target
```

```python
wine.feature_names
```

```python
df = pd.DataFrame(wine.data, columns = wine.feature_names)
df.head()
```

```python
df = pd.concat([df, pd.DataFrame(wine.target, columns = ['Categories'])], axis = 'columns')
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
inputs = df.drop('Categories',axis = 'columns')
```

```python
y = df[['Categories']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, y, test_size = 0.2, random_state = 2)
```

### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
y_test
```

### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
